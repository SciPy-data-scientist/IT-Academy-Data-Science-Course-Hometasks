Отмечаем узкое распределение данных (напоминает экспоненциальное распределение) практически для всех показателей (features) в сочетании с большим количеством выбросов. Практически для всех показателей минимальные значения равняются 0, т.е. данные скошены вправо. Большое количество выбросов может оказаться проблемой при построении моделей машинного обучения, которые чувствительны к выбросам. Предполагается в ходе последующей предобработки данных заменить те выбросы, которые превышают 95-й процентиль, значением 95-го квантиля (Почему именно так? Потому что, чаще всего, выбросы в данном датасете, имеющем сдвиг вправо, представляют собой значения из области максимальных значений. Заменяя их медианой, мы заменяем их значениями из области минимальных значений (часто медиана равняется 0), что может изменить распределение данных. В связи с этим было принято решение заменить выбросы значениями 95-го квантиля. Тем не менее, в некоторых случаях, как, например, для столбца "drop_dat_Mean" 95-й квантиль соответствует медиане, т.е. после замены будет получен столбец с одинаковыми значениями признака. Такие столбцы, поскольку они не несут никакой информации для модели машинного обучения, будут удалены, что позволит снизить размерность данных). Кроме того, вместо замены выбросов медианой, можно использовать для стандартизации RobustScaler, а не StandardScaler или MinMaxScaler, поскольку последние два чувствительны к выбросам и при наличии сильно отклоняющихся выбросов приводят к сжатию данных в узком диапазоне. RobustScaler не чувствителен к выбросам и приводит к стандартизации данных внутри интервала IQR.

Также отмечаем наличие отрицательных минимальных значений для столбцов "rev_Mean" и "totmrc_Mean", а также "change_rev" и "change_mou". Для второй пары столбцов такие значения понятны. Для первой пары отрицательные значения кажутся странными. Возможно, это отрицательный баланс на начало и конец отчетного периода для конкретного пользователя. Для остальных показателей значения соответствуют логике: минимальные значения представляют собой неотрицательные значения.

Что касается значений nan, то в выбранных столбцах их количество невелико. Работа с такими значениями предполагается в конце EDA. Однако в общем датасете df можно отметить наличие строк с большим количеством (>10) значений nan. Предполагается такие строки удалить из датасета, исходя из той логики, что, если для таких наблюдений мы выполним подстановку медианного (либо любого иного) значения, мы внесем слишком большое количество изменений (т.е. для таких наблюдений более 10 % значений будут искусственно созданы, данные окажутся отчасти "синтезированными").

Создадим функцию для замены отсутствующих данных (nan) в категориальных признаках на "Unknown", если их значение превышает 5 % от всего количества данных в столбце, либо на моду, если их значение не превышает либо равно 5 % от всего количества данных в столбце:

```python
def replace_na_mode_unknown(df):
    for column in df.select_dtypes(include=['object']).columns:
        nan_count = df[column].isna().sum()
        total_count = len(df[column])
        nan_percentage = nan_count / total_count

        if nan_percentage > 0.05:
            df[column] = df[column].fillna('Unknown')
        else:
            mode_value = df[column].mode()[0]
            df[column] = df[column].fillna(mode_value)
    return df
```

Создадим функцию для определения категорий признака, для которых накопленная сумма частот близка к 95 % (аналог 95-го процентиля), и присвоения всем остальным категориям значения "other".

```python
def top_95prc_categories(df, column):
    category_counts = df[column].value_counts(normalize=True).cumsum()
    top_95prc_categories = category_counts[category_counts <= 0.95].index
    df[column] = df[column].apply(lambda x: x if x in top_95prc_categories else 'other')
```

Создадим функцию для замены значений nan в численных столбцах на медиану значений соответствующего столбца:

```python
def replace_nan_median(df, columns):
    for column in columns:
        median_value = df[column].median()
        df[column] = df[column].fillna(median_value)
    return df
```

Создадим функцию для замены значений nan в столбцах с упорядоченными категориальными данными на категориальные значения, которые наблюдаются для данного признака, с сохранением исходного распределения данных, т.е. если значение 0 встречалось 50 %, то при замене значений nan значение 0 будет подставлено в 50 % случаев:

```python
def replace_nan_random_value(df, columns):
    for column in columns:
        value_counts = df[column].value_counts(normalize=True)
        nan_count = df[column].isna().sum()
        fill_values = np.random.choice(value_counts.index, size=nan_count, p=value_counts.values)
        df.loc[df[column].isna(), column] = fill_values
    return df
```

#### 1.7 Замена выбросов для столбцов с непрерывными числовыми и упорядоченными категориальными данными на значения 95-го квантиля

```python
def outlier_detect(data,column,quantile=0.95):
    fence = np.quantile(data[column], quantile)
    tmp = pd.DataFrame(data[column] > fence)
    outliers = tmp.any(axis=1)
    outlier_indices = data[outliers].index

    return outlier_indices
```

Потенциально это может ухудшить качество модели, однако применение OrdinalEncoder вносит упорядочение в категориальные данные, которое изначально отсутствовало, что также имеет негативное влияние. Тем не менее, можно провести сравнение обоих методов для определения наилучшего варианта кодирования категориальных данных.

Упорядоченные категориальные данные по сути уже закодированы и не требуют дополнительной обработки.

Во избежание утечки данных в модель разделим датасет на train и тест выборку. Выполним fit_transform для train данных и tranform для test данных.

```python
df.info()
```

```python
X = df.drop("churn", axis=1)
y = df["churn"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

Выполним кодировку категориальных данных, выбрав их в соответствии с типом данных "object". После этого объединим закодированные категориальные данные с числовыми данными.

```python
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

X_train_encoded = encoder.fit_transform(X_train.select_dtypes("object"))
X_train_encoded_df = pd.DataFrame(X_train_encoded, columns=encoder.get_feature_names_out(X_train.select_dtypes("object").columns))

X_train_final = pd.concat([X_train_encoded_df.reset_index(drop=True), X_train.select_dtypes(include=["float", "int"]).reset_index(drop=True)], axis=1)

print("Train данные после OneHotEncoding:")
print(X_train_final.shape)

X_test_encoded = encoder.transform(X_test.select_dtypes("object"))
X_test_encoded_df = pd.DataFrame(X_test_encoded, columns=encoder.get_feature_names_out(X_test.select_dtypes("object").columns))

X_test_final = pd.concat([X_test_encoded_df.reset_index(drop=True), X_test.select_dtypes(include=["float", "int"]).reset_index(drop=True)], axis=1)

print("Test данные после OneHotEncoding:")
print(X_test_final.shape)
```

# 2.2 Шкалирование численных данных с помощью RobustScaler

Шкалирование с помощью StandardScaler/RobustScaler следует проводить только для непрерывных численных данных, исключая упорядоченные категориальные данные, выраженные числами. Поэтому необходимо отделить непрерывные численные данные от упорядоченных категориальных данных. Однозначно это сделать невозможно, глядя на весь датасет в целом. Однако один из способов разделения данных - это разделение на основе количества уникальных значений в соответствующих столбцах. Для упорядоченных категориальных данных число уникальных значений будет относительно небольшим (в данном случае пороговое значения принято равным 20). При наличии более 20 уникальных значений в столбце примем, что это непрерывные численные данные.

```python
threshold = 20
potential_ordinal_columns = [col for col in X.select_dtypes(include=["float", "int"]).columns if X[col].nunique() <= threshold]

true_numerical_columns = [col for col in X.select_dtypes(include=["float", "int"]).columns if col not in potential_ordinal_columns]
```

Для train и test выборок выделим столбцы с непрерывными численными данными. Все столбцы, не содержащие непрерывных численных данных, отнесем к остальным. Проведем шкалирование для непрерывных численных данных, после чего объединим полученный промежуточный датасет с остальными (отложенными) столбцами, которые не подвергали шкалированию.

```python
X_train_true_numerical_features = X_train_final[true_numerical_columns]
X_train_others = X_train_final.drop(true_numerical_columns, axis=1)

#scaler = StandardScaler()
scaler = RobustScaler()
X_train_scaled = scaler.fit_transform(X_train_true_numerical_features)
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=X_train_true_numerical_features.columns)

X_train_transformed = pd.concat([X_train_scaled_df.reset_index(drop=True), X_train_others.reset_index(drop=True)], axis=1)


X_test_true_numerical_features = X_test_final[true_numerical_columns]
X_test_others = X_test_final.drop(true_numerical_columns, axis=1)

X_test_scaled = scaler.transform(X_test_true_numerical_features)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=X_test_true_numerical_features.columns)

X_test_transformed = pd.concat([X_test_scaled_df.reset_index(drop=True), X_test_others.reset_index(drop=True)], axis=1)


print("Transformed Train Data:\n", X_train_transformed.shape)
print("\nTransformed Test Data:\n", X_test_transformed.shape)
```

### 3.1 Отбор признаков для модели на основе LogisticRegression, построение соответствующей модели и оценка ее качества

Для отбора признаков использовали данные после кодирования с помощью OneHotEncoder и шкалирования с помощью StandardScaler или RobustScaler. После отбора наиболее значимых признаков построим с их помощью соответствующую модель и оценим ее качества. Кроме того, в целях сравнения построим модель на всем пространстве признаков. Для отбора признаков использовал SequentialFeatureSelector, который может работать как с численными, так и категориальными признаками. В данном случае применяли метод "forward", который предполагает поэтапное добавление признаков в исследуемую модель и оценку метрики качества (accuracy). Намеренно не указывал, какое количество признаков следует отобрать (n_features_to_select="auto"), но указал предел, когда следует остановить отбор (tol=0.001).

```python
sfs_lr_forward = SequentialFeatureSelector(
    LogisticRegression(), n_features_to_select="auto", tol=0.001, direction="forward", cv=5, n_jobs=-1
    ).fit(X_train_transformed, y_train)

log_reg_col_selected = X_train_transformed.columns[sfs_lr_forward.get_support()]

print(
    "Features selected by forward sequential selection: \n"
    f"{log_reg_col_selected}"
)
```

# 5. Feature Engineering

Поскольку множество числовых данных имеют экспоненциальное распределение, то, возможно, полезным будет логарифмическое преобразование числовых данных с большими правыми хвостами. Для преобразования будем использовать выборки, полученные после разбиения на train и test, т.е. X_train и X_test. Кроме того, из них необходимо выбрать только те признаки, которые принимают непрерывные числовые значения.

```python
print(X_train.shape, X_test.shape)
```

Как и ранее для шкалирования данных, выделем непрерывные числовые признаки и упорядоченные категориальные признаки, выраженные числами.

```python
threshold = 20
potential_ordinal_columns = [col for col in X.select_dtypes(include=["float", "int"]).columns if X[col].nunique() <= threshold]

true_numerical_columns = [col for col in X.select_dtypes(include=["float", "int"]).columns if col not in potential_ordinal_columns]
```

```python
X_train_true_numerical_features = X_train[true_numerical_columns]
X_train_others = X_train.drop(true_numerical_columns, axis=1)

X_test_true_numerical_features = X_test[true_numerical_columns]
X_test_others = X_test.drop(true_numerical_columns, axis=1)
```

Выполним преобразование данных с помощью трансформера PowerTransformer, используя метод 'yeo-johnson'. Этот трансформер выполняет логарифмическое преобразование данных, однако отличие от логарифмической функции может работать с отрицательными и нулевыми значениями, которых много в датасете.

```python
#transformer = PolynomialFeatures(degree=2)
transformer = PowerTransformer(method='yeo-johnson')
```

```python
X_train_log_transformed = transformer.fit_transform(X_train_true_numerical_features)

X_train_log_transformed_df = pd.DataFrame(X_train_log_transformed, columns=transformer.get_feature_names_out(X_train_true_numerical_features.columns))
X_train_log = pd.concat([X_train_log_transformed_df.reset_index(drop=True), X_train_others.reset_index(drop=True)], axis=1)


X_test_log_transformed = transformer.transform(X_test_true_numerical_features)
X_test_log_transformed_df = pd.DataFrame(X_test_log_transformed, columns=transformer.get_feature_names_out(X_test_true_numerical_features.columns))

X_test_log = pd.concat([X_test_log_transformed_df.reset_index(drop=True), X_test_others.reset_index(drop=True)], axis=1)
```

```python
print(X_train_log.shape, X_test_log.shape)
```

Выполнив логарифмическое преобразование данных, мы можем осуществить все те же шаги, которые выполняли ранее для непребразованного датасета, т.е. кодирование с помощью OneHotEncoder и шкалирование с помощью RobustScaler, отбор признаков с помощью SequentialFeatureSelector с последующим построением модели машинного обучения на выбранных признаках и подбором гиперпараметров.

Кодирование категориальных данных с помощью OneHotEncoder:

```python
encoder = OneHotEncoder(handle_unknown='ignore', sparse_output=False)

X_train_log_encoded = encoder.fit_transform(X_train_log.select_dtypes("object"))
X_train_log_encoded_df = pd.DataFrame(X_train_log_encoded, columns=encoder.get_feature_names_out(X_train_log.select_dtypes("object").columns))

X_train_log_final = pd.concat([X_train_log_encoded_df.reset_index(drop=True), X_train_log.select_dtypes(include=["float", "int"]).reset_index(drop=True)], axis=1)

print("Train данные после OneHotEncoding:")
print(X_train_log_final.shape)

X_test_log_encoded = encoder.transform(X_test_log.select_dtypes("object"))
X_test_log_encoded_df = pd.DataFrame(X_test_log_encoded, columns=encoder.get_feature_names_out(X_test_log.select_dtypes("object").columns))

X_test_log_final = pd.concat([X_test_log_encoded_df.reset_index(drop=True), X_test_log.select_dtypes(include=["float", "int"]).reset_index(drop=True)], axis=1)

print("Test данные после OneHotEncoding:")
print(X_test_log_final.shape)
```

Шкалирование данных с помощью RobustScaler: 

```python
X_train_log_true_numerical_features = X_train_log_final[true_numerical_columns]
X_train_log_others = X_train_log_final.drop(true_numerical_columns, axis=1)

scaler = MinMaxScaler()
X_train_log_scaled = scaler.fit_transform(X_train_log_true_numerical_features)
X_train_log_scaled_df = pd.DataFrame(X_train_log_scaled, columns=X_train_log_true_numerical_features.columns)

X_train_log_transformed = pd.concat([X_train_log_scaled_df.reset_index(drop=True), X_train_log_others.reset_index(drop=True)], axis=1)


X_test_log_true_numerical_features = X_test_log_final[true_numerical_columns]
X_test_log_others = X_test_log_final.drop(true_numerical_columns, axis=1)

X_test_log_scaled = scaler.transform(X_test_log_true_numerical_features)
X_test_log_scaled_df = pd.DataFrame(X_test_log_scaled, columns=X_test_log_true_numerical_features.columns)

X_test_log_transformed = pd.concat([X_test_log_scaled_df.reset_index(drop=True), X_test_log_others.reset_index(drop=True)], axis=1)


print("Transformed Train Data:\n", X_train_log_transformed.shape)
print("\nTransformed Test Data:\n", X_test_log_transformed.shape)
```
